# Поиск мотивов через HMMER

### Загрузка геномов
Mesocricetus auratus (golden hamster): datasets download genome accession GCF_017639785.1 --include protein


Rattus norvegicus (Norway rat): datasets download genome accession GCF_036323735.1 --include protein

### HMM
Взяла PF00004 ATPase family associated with various cellular activities (AAA)

AAA family proteins often perform chaperone-like functions that assist in the assembly, operation, or disassembly of protein complexes [1].

wget https://www.ebi.ac.uk/interpro/wwwapi//entry/pfam/PF00004?annotation=hmm -O PF00271.hmm.gz

Затем запускаем поиск доменов HMMER:

hmmsearch --tblout hamster_hmmsearch.tbl PF00271.hmm hamster_dataset/data/GCF_017639785.1/protein.faa > hamster_hmmsearch.txt

hmmsearch --tblout rattus_hmmsearch.tbl PF00271.hmm rattus/data/GCF_036323735.1/protein.faa > rattus_hmmsearch.txt


Возьмём 2 белка с наименьшими e-value из каждого файла и извлечём эти белки из FASTA-файлов


seqkit grep -f ids.txt all_proteins.faa > top4_proteins.fasta


### Выравнивание
Используем clustalo -i top4_proteins.fasta -o aligned_top4.aln --outfmt=clu
