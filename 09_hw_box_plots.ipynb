```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Сгенерируем данные из нормального распределения и построим ящик с усами
data = np.random.uniform(low=0, high=10, size=100)

plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True)
plt.title('Ящик с усами для равномерного распределения')
plt.ylabel('Значения')
plt.grid(True, alpha=0.7)
plt.show()
```

```python
data = [79, 80, 80, 80, 74, 80, 80, 79, 64, 78, 73, 78, 74, 45, 81, 48, 80, 82, 82, 70]

#(a.) Получите квартили.
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)

print(f"Q1 (25-й перцентиль) = {Q1}")
print(f"Q2 (медиана) = {Q2}")
print(f"Q3 (75-й перцентиль) = {Q3}")
```

```python
# (b.) Определите межквартильный размах. 
IQR = Q3 - Q1
print(f"Межквартильный размах (IQR) = {IQR}")
```

```python
# (c.) Найдите минимум, Q25, медиану, Q75 и максимум 
mini = np.min(data)
maxi = np.max(data)

print(f"Минимум = {mini}")
print(f"Q1 (25-й перцентиль) = {Q1}")
print(f"Q2 (медиана) = {Q2}")
print(f"Q3 (75-й перцентиль) = {Q3}")
print(f"Максимум = {maxi}")
```

```python
# (d.) Определите потенциальные выбросы, если таковые имеются. 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Границы выбросов: [{lower_bound}, {upper_bound}]")
print(f"Потенциальные выбросы: {outliers}")
```

```python
# (e.) Постройте ящик с усами.
plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True)
plt.title('Ящик с усами для равномерного распределения')
plt.ylabel('Значения')
plt.grid(True, alpha=0.7)
plt.show()
```
