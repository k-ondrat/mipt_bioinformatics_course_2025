```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Чтение таблиц

```python
data = pd.read_csv('~/homework/hw_2/TRB_10_7/TRB_10_7.tsv', sep='\t')
data
```

```python
# Сделаем более user-friendly названия столбцов
data.columns = ['nucl_seq', 'prot_seq', 'TRBV', 'TRBJ']
data
```

## Построение графика

```python
# Сложно придумать информационно значимый график для данных без цифр...
value_counts = data['TRBV'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(value_counts.index, value_counts.values, color='skyblue')

plt.xticks(rotation=90)  # Поворачиваем подписи оси X для удобства чтения
plt.xlabel("Тип TRBV")
plt.ylabel("Частота встречаемости")
plt.title("Гистограмма встречаемости значений в TRBV")

plt.show()
```

## Передадим датафрейм в среду R

```python
data_r = pandas2ri.py2rpy(data[:1000000]) #пришлось укоротить, а то ядро умирает от 10 млн строк...
```

```python
from rpy2 import robjects 
robjects.globalenv['data_r'] = data_r #Перенесём наши данные в среду R
```

## Построим аналогичный график для TRBJ

```python
%%R
ggplot(data_r, aes(x = TRBJ)) + 
geom_bar(fill = 'pink') +
labs(title = "Гистограмма встречаемости значений в TRBJ",
       x = "Тип TRBJ",
       y = "Частота") +
theme(plot.title = element_text(hjust = 0.5), #центрируем название
      axis.text.x = element_text(angle = 90))
```
